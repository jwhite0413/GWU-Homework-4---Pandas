```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
heroes = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
heroes_data_df = pd.read_csv(heroes)
heroes_data_df
```

## Player Count

* Display the total number of players


```python
#PlayerCount
player_count = len(heroes_data_df["SN"].unique())
player_count

#totalplayersdisplay
total_players_df = pd.DataFrame({"Total Players":[player_count]})
total_players_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#item count
items = len(heroes_data_df["Item ID"].unique())
items

#average price 
average_price = round(heroes_data_df["Price"].mean(),2)
average_price

#number of purchases
purchases = len(heroes_data_df["Item ID"])
purchases

#Total revenue
revenue = round(sum(heroes_data_df["Price"]),2)
revenue
```

```python
#dataframe for the above data
purchase_analysis_df = pd.DataFrame({"Number of Unique Items":[items],"Average Price":[average_price],"Number of Purchases":[purchases],"Total Revenue":[revenue]})
purchase_analysis_df

purchase_analysis_df["Average Price"] = purchase_analysis_df["Average Price"].map("${:.2f}".format)
purchase_analysis_df["Total Revenue"] = purchase_analysis_df["Total Revenue"].map("${:,.2f}".format)
purchase_analysis_df
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#dataframe for unique SNs and gender

# Purchase ID	SN	Age	Gender	Item ID	Item Name	Price

new_df =pd.DataFrame(heroes_data_df,columns=["Purchase ID","SN" ,"Gender","Age","Item ID","Price" ])


new_df
```

```python
test = new_df.drop_duplicates(subset ="SN", keep = 'first', inplace = True) 
test

male_female_count = new_df["Gender"].value_counts()
male_female_count

count_df = pd.DataFrame(male_female_count)
count_df




total_count = sum(male_female_count)
total_count


male_female_df =pd.DataFrame (male_female_count,columns=["Gender"])





male_female_df["Total Percent"] = ((male_female_df['Gender'] / total_count)*100).round(2)
male_female_df

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase = heroes_data_df.groupby(["Gender"]).count()["Price"]
avg_price = heroes_data_df.groupby(["Gender"]).mean()["Price"]
total = heroes_data_df.groupby(["Gender"]).sum()["Price"]
avg_total = total/male_female_count


analysis_df = pd.DataFrame ({"Purchase Count":purchase,"Average Purchase Price":avg_price,"Total Purchase Value":total, "Average Per Person":avg_total})
analysis_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9,14,19,24,29,34,39,100]
groups = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
new_df ["Age Ranges"] = pd.cut(new_df["Age"],bins, labels = groups)
new_df

age_total = new_df["Age Ranges"].value_counts()
age_total
age_percent = (age_total/total_count)
age_percent
demo_table = pd.DataFrame ({"Total":age_total,"Percent of Total":age_percent})
demo_table["Percent of Total"] = demo_table ["Percent of Total"].map("{:,.2%}".format)
demo_table = demo_table.reindex(["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"])
demo_table
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0,9,14,19,24,29,34,39,100]
groups = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]
# age_analysis = new_df.loc[:,["SN","Age","Price"]]
heroes_data_df["Age Ranges"] = pd.cut(heroes_data_df["Age"],bins, labels = groups)

new_df
# age_total = age_analysis.sum()

# age_total = age_analysis["Age Ranges"].value_counts()
# age_total

purchase_age = heroes_data_df.groupby(["Age Ranges"]).count()["Age"]
purchase_age
avg_price_age = heroes_data_df.groupby(["Age Ranges"]).mean()["Price"]
total_age = heroes_data_df.groupby(["Age Ranges"]).sum()["Price"]
avg_total_age = total_age/purchase_age


age_analysis_df = pd.DataFrame ({"Purchase Count":purchase_age,"Average Purchase Price":avg_price_age,"Total Purchase Value":total_age, "Average Per Person":avg_total_age})
age_analysis_df


```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
group_by_sn = heroes_data_df.groupby(["SN"])
group_by_sn

group_price = group_by_sn.sum()["Price"]
group_price

average_price = group_by_sn.mean()["Price"]
average_price

group_count = group_by_sn.count()["Price"]
group_count

top_df = pd.DataFrame ({"Purchase Count":group_count,"Average Purchase Price":average_price,"Total Purchase Price":group_price})
top_df

group_sort_df = top_df.sort_values("Total Purchase Price", ascending = False)
group_sort_df.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular = heroes_data_df.groupby(["Item ID","Item Name"])
popular


popular_total = popular.sum()["Price"]
popular_total

popular_average = popular.mean()["Price"]
popular_average

popular_count = popular.count()["Price"]


popular_items_df = pd.DataFrame ({"Count": popular_count,"Average Purchase Price": popular_average,"Total Purchase Price": popular_total})

sort_popular = popular_items_df.sort_values("Count", ascending = False)

sort_popular.head(5)

```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
sort_value = popular_items_df.sort_values("Total Purchase Price",ascending = False)
sort_value.head()
```
